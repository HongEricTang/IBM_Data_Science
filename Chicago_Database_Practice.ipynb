##### How many crimes involve an arrest?

```python
answer = pd.read_sql('SELECT count(*) as "Crimes Involving Arrest" FROM chicago_crime where arrest = "TRUE"', con=mydb)
answer
```

##### Which unique types of crimes have been recorded at GAS STATION locations?


```python
answer = pd.read_sql('SELECT distinct(primary_type) as "Unique Types of Crime" FROM chicago_crime', con=mydb)
answer
```

##### In the CENUS_DATA table list all Community Areas whose names start with the letter ‘B’.

```python
answer = pd.read_sql('SELECT * FROM census where community_area_name like "B%"', con=mydb)
answer
```

##### Which schools in Community Areas 10 to 15 are healthy school certified?

```python
answer = pd.read_sql('SELECT * FROM chicago_public_schools where HEALTHY_SCHOOL_CERTIFIED = "Yes" and (Community_area_number between 10 and 15)', con=mydb)
answer
```

##### What is the average school Safety Score? 

```python
answer = pd.read_sql('SELECT AVG(safety_score) as "Average School Safety Score" FROM chicago_public_schools', con=mydb)
answer
```

##### List the top 5 Community Areas by average College Enrollment [number of students] 

```python
answer = pd.read_sql(
    'select * from (SELECT COMMUNITY_AREA_NAME as "Community", sum(college_enrollment) as "Total" FROM chicago_public_schools group by COMMUNITY_AREA_NAME) as tmp order by Total desc limit 5'
    , con=mydb)
answer
```

##### Use a sub-query to determine which Community Area has the least value for school Safety Score? 

```python
answer = pd.read_sql(
    'select * from (SELECT COMMUNITY_AREA_NAME as "Community", avg(safety_score) as "Average" FROM chicago_public_schools group by COMMUNITY_AREA_NAME) as tmp where average = (select min(Average) from (SELECT COMMUNITY_AREA_NAME as "Community", avg(safety_score) as "Average" FROM chicago_public_schools group by COMMUNITY_AREA_NAME) as tmp)'
    , con=mydb)
answer
```

##### Find the Per Capita Income of the Community Area which has a school Safety Score of 1.

```python
answer = pd.read_sql(
    'select community_area_name, per_capita_income from census where community_area_name in (select community_area_name from chicago_public_schools where safety_score = 1)'
    , con=mydb)
answer
```
